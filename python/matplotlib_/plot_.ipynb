```python
import matplotlib.pyplot as plt
import numpy as np
```

### データの作成

```python
x  = np.linspace(0, 10, 101) # [0., 10.] の区間から等間隔で11個の点を作成 → 0., 1., 2., ... , 10.
y1 = np.sin(x)
y2 = 10 * np.cos(x)
```

```python
height = 5
width  = 5
legendsize = 10
labelsize  = 13
ticksize   = 14


# fig : グラフが描画されるキャンパス、ax : axes=グラフオブジェクト
# 1つのキャンパスに複数のグラフ領域を作るのはめんどくさいのでやらない
fig, ax1 = plt.subplots(figsize=(width,height))
# 第2軸の定義 ： 同一のグラフの中に2つ目の縦軸を定義
ax2 = plt.twinx()

ax1.plot(x, y1, label='サイン関数', color='g', linewidth=10, alpha=0.1, linestyle='-')
ax2.plot(x, y2, label='コサイン関数',  color='b', linewidth=5,  alpha=0.8, linestyle=':')

# 軸のラベルの設定
# WindowsならMeiryoで日本語が使える。他は知らない
ax1.set_xlabel("x軸", fontname="Hiragino Sans", fontsize=labelsize)
ax1.set_ylabel("y軸１", fontname="Hiragino Sans", fontsize=labelsize)
ax2.set_ylabel("y軸２", fontname="Hiragino Sans", fontsize=labelsize)
# twinx()に対するset_xlabelは何の効果もないっぽい → エラーも出ないし、設定も変更等されない
# ax2.set_xlabel("x2軸", fontname="Hiragino Sans", fontsize=30)


# 軸のメモリの設定
ax1.tick_params(axis='x', labelsize=ticksize)
ax1.tick_params(axis='y', labelsize=ticksize-2)
ax2.tick_params(axis='y', labelsize=ticksize+3)

fig.legend(prop={'family':"Hiragino Sans", 'size': legendsize})
```
